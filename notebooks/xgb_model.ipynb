```python
y = df['Churn']
X = df.drop('Churn', axis=1)

print(f"\nDimensions of X (features): {X.shape}")
print(f"Dimension of y (alvo): {y.shape}")
```

# 2. Machine Learning - XGBoost

## 2.1 Pipeline and Model Train

```python
# Calculate counts for each class
class_counts = y_train.value_counts()
count_negative = class_counts[0] # Count of 'No Churn'
count_positive = class_counts[1] # Count of 'Yes Churn'

# Calculate the scale_pos_weight
scale_pos_weight_value = count_negative / count_positive

print(f"Negative (0) samples in y_train: {count_negative}")
print(f"Positive (1) samples in y_train: {count_positive}")
print(f"Calculated scale_pos_weight: {scale_pos_weight_value:.4f}")
```

```python
# Pipeline for XGBoost
model_xgb = Pipeline([
    ('model', XGBClassifier(
        random_state=42,
        scale_pos_weight=scale_pos_weight_value,  
        n_jobs=-1,
        eval_metric='logloss'     
    ))
])

model_xgb.fit(X_train, y_train)
```

```python
correct_features = list(X_train.columns)

# Number of features the model was trained on
n_features_in_model = model_xgb.n_features_in_

# Number of feature names you provided
n_features_in_list = len(correct_features)

print(f"Model expects {n_features_in_model} features.")
print(f"You provided {n_features_in_list} feature names.")

# This condition is likely True in your case
if n_features_in_model != n_features_in_list:
    print("\n⚠️ There is a mismatch, which is causing the error.")
```

## 2.2 Cross-Validation and Hyperparameter Tuning

```python
def perform_grid_search_xgb(X_train_data, y_train_data, param_grid, scale_pos_weight):
    """
    Performs GridSearchCV to find the best hyperparameters for an XGBoost model.
    Uses scale_pos_weight for imbalanced data.
    Returns training and validation scores for the best model.

    Args:
        X_train_data (pd.DataFrame): Training features (unscaled).
        y_train_data (pd.Series): Training labels.
        param_grid (dict): Dictionary with hyperparameters.
        scale_pos_weight (float): Pre-calculated weight for the positive class.

    Returns:
        xgboost.XGBClassifier: The best model found by GridSearchCV.
        dict: The best hyperparameters.
        float: The best cross-validation score (validation score).
        float: The corresponding training score for the best model.
    """
    
    # Initialize the XGBoost model
    xgb_model = XGBClassifier(
        random_state=42,
        scale_pos_weight=scale_pos_weight,  # Handle class imbalance
        n_jobs=-1,                          # Use all cores
        eval_metric='logloss'               # Suppress warnings
    )
    
    
    grid_search = GridSearchCV(
        estimator=xgb_model, # Use the XGBoost model
        param_grid=param_grid, 
        cv=5, 
        scoring='recall', 
        return_train_score=True,
        n_jobs=-1  # Use all cores for the grid search itself
    )
    
    # Fit the grid search
    grid_search.fit(X_train_data, y_train_data)
    
    
    # Get the scores for the best model
    best_model_index = grid_search.best_index_
    best_train_score = grid_search.cv_results_['mean_train_score'][best_model_index]
    
    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_, best_train_score
```

```python
# Hyperparameter Grid Definition for XGBoost


param_grid_xgb = {
    'n_estimators': [100, 150],
    'max_depth': [5, 7],
    'learning_rate': [0.05, 0.1],
    'subsample': [0.7, 0.8],
    'colsample_bytree': [0.7, 0.8]
}

print("Starting GridSearchCV for XGBoost (this may take a while)...")

# Call the Grid Search Function
best_model_xgb, best_hyperparams_xgb, best_validation_score_xgb, best_train_score_xgb = perform_grid_search_xgb(
    X_train, y_train, param_grid_xgb, scale_pos_weight_value
)

# Results Analysis
print("\n--- XGBoost Results ---")
print(f"Best Hyperparameters: {best_hyperparams_xgb}")
print(f"Best Validation Recall (CV): {best_validation_score_xgb:.4f}")
print(f"Training Recall (for the best model): {best_train_score_xgb:.4f}")
```
